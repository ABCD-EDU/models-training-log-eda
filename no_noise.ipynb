```python
import matplotlib.pyplot as plt

train_sisnr = [-1.41, -3.02, -3.82, -4.56, -5.10, -5.59, -5.89, -6.37, -6.64, -6.98,
               -7.31, -7.43, -7.77, -7.94, -8.15, -8.47, -8.73, -8.92, -8.78, -8.67,
               -9.43, -9.63, -9.85, -9.95, -10.2, -10.4, -10.4, -10.6, -10.6]

valid_sisnr = [-2.13, -3.25, -4.20, -4.96, -5.34, -5.78, -6.00, -6.28, -6.69, -6.81,
               -6.72, -7.08, -7.81, -7.69, -8.02, -8.33, -8.58, -8.41, -8.10, -8.35,
               -9.31, -9.32, -9.50, -9.73, -9.90, -10.1, -10.1, -9.98, -9.98]

epochs = range(1, 30)
lr = [1.50e-04] * 20 + [7.50e-05] * 9

# plot training and validation si-snr vs epochs
fig, ax = plt.subplots(figsize=(20, 13))
ax.plot(epochs, train_sisnr, 'b', label='Train si-snr')
ax.plot(epochs, valid_sisnr, 'r', label='Valid si-snr')
ax.set_xlabel('Epoch')
ax.set_ylabel('si-snr')
ax.set_title('Training and validation si-snr vs epochs')
ax.legend()

# create a new figure and plot learning rate vs epochs
fig2, ax2 = plt.subplots(figsize=(20, 13))
ax2.plot(epochs, lr, 'g', label='Learning rate')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Learning rate')
ax2.set_title('Learning rate vs epochs')
ax2.legend()

plt.show()
```
